```python
!install pandas

import pandas as pd

df_path = "C:\\Users\\alexa\\OneDrive\\Área de Trabalho\\Area Tech\\Base Exercicio (em milhoes).csv"
df = pd.read_csv(df_path)

df[:5]
```

```python
def calcula_dre():
    for i in range(len(df)):
        faturamento = df['volume_alunos'][i] * df['preco'][i]
        lucro = faturamento - df['despesas_custos'][i] - df['ir'][i]
        df.iat[i,df.columns.get_loc('faturamento')] = faturamento
        df.iat[i,df.columns.get_loc('lucro_liquido')] = lucro
        df.to_excel('Tech - Gabarito Aula 1.xlsx', index=False)
    return df
```

```python
calcula_dre()
```

```python
path = "C:\\Users\\alexa\\OneDrive\\Área de Trabalho\\base_ibov.csv"
df = pd.read_csv(path)
print(df)

def perform_ibov():
    for i in range(len(df)):
        if df['ebitda'][i] > 30000000:
            df.iat[i,df.columns.get_loc('perform')] = 'alta' 
        elif df['ebitda'][i] > 20000000 and df['ebitda'][i] < 30000000:
            df.iat[i,df.columns.get_loc('perform')] = 'media'
        else:
            df.iat[i,df.columns.get_loc('perform')] = 'baixa'
        i += 1
    df.to_excel('ebitda_perfomance.xlsx', index=False)
    return df
```

```python
perform_ibov()
```
