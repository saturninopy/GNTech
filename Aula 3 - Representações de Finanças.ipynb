```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import yfinance as yf

ativos = ['EMBR3.SA','PETZ3.SA','JBSS3.SA','VIVA3.SA','NTCO3.SA','VALE3.SA']

precos = pd.DataFrame()

for i in ativos:
    precos[i] = yf.download(i, start = '2022-01-01', end = '2022-04-20')['Adj Close']
```

```python
ativos_retornos = precos.pct_change()
ativos_retornos = ativos_retornos.dropna()
ativos_retornos
```

```python
volt = ativos_retornos.std()

volatilidade = pd.DataFrame(ativos_retornos.std(), columns = ['Volatility'])

volt
```

```python
retn = ativos_retornos.mean()

retornos_medios = pd.DataFrame(ativos_retornos.mean(), columns = ['Returns'])

retn
```

```python
risco_retorno = pd.concat([retornos_medios, volatilidade], axis = 1)

risco_retorno 
```

```python
risco_retorno['Sharpe'] = risco_retorno['Returns']/risco_retorno['Volatility']
risco_retorno
```

```python
plt.subplots(figsize = (8,6))

sns.scatterplot(data = risco_retorno, x = 'Volatility', y = 'Returns')

for i in range(risco_retorno.shape[0]):
    plt.text(x = risco_retorno.Volatility[i], y = risco_retorno.Returns[i],
             s = risco_retorno.index[i],
             fontdict = dict(color = 'black', size = 8),
             bbox = dict(facecolor = 'lime'))
```
