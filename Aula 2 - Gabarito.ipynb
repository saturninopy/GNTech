```python
# REPORT - GNASSET

import quantstats as qs

qs.extend_pandas()

tickers = {
    'FLRY3.SA': 0.12,
    'RENT3.SA': 0.12,
    'CRFB3.SA': 0.12,
  # 'ABEV3.SA': 0.1,
    'TOTS3.SA': 0.12,
  #2 'BPAC11.SA': 0.1, 
  #3 'ENJU3.SA': 0.1,
    'RACE': 0.12,
  # 'SHOP': 0.133,
    'HOME34.SA': 0.155,
  # 'ANIM3.SA': 0.12, 
  # 'HYPE3.SA': 0.12
}

gnasset = qs.utils.make_index(tickers)

qs.reports.html(gnasset, 'SPY', title='GNAsset vs. S&P500', output='GNAsset_vs_S&P.html')

qs.plots.earnings(gnasset,start_balance=10000)
```
