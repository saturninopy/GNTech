```python
# REPORT - Fundo ESG

import quantstats as qs

qs.extend_pandas()

tickers = {
    'EMBR3.SA': 0.296,
    'PETZ3.SA': 0.569, 
    'JBSS3.SA': 0.053,
    'VIVA3.SA': 0.048,
    'NTCO3.SA': 0.035
}

gnasset = qs.utils.make_index(tickers)

qs.reports.html(gnasset, '^BVSP', title='Moderninho Investments vs. IBOV', output='Moderninho_vs_S&P.html')

qs.plots.earnings(gnasset,start_balance=100000)
```
