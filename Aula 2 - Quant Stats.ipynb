```python
# SINGLE STOCK - SANTOS BRASIL

import quantstats as qs

qs.extend_pandas()

stock = qs.utils.download_returns("STBP3.SA")

# print('CAGR: ', stock.cagr())

sharpe_ratio = qs.stats.sharpe(stock)

print('Sharpe Ratio: ', sharpe_ratio)

print('Max Drawdown: ', stock.max_drawdown())

qs.plots.earnings(stock,start_balance=10000)

stock.plot_monthly_heatmap(savefig='fb_monthly_heatmap.png')
```

```python
# REPORT - GNASSET

import quantstats as qs

qs.extend_pandas()

tickers = {
    'FB': 0.2,
    'AAPL': 0.2,
    'AMZN': 0.2,
    'NFLX': 0.2,
    'GOOG': 0.2,
}

faang = qs.utils.make_index(tickers)

qs.reports.html(faang, 'SPY', title='FAANG vs. S&P500', output='FAANG_vs_S&P.html')

qs.plots.earnings(faang,start_balance=10000)
```
